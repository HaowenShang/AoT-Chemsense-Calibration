```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Get the raw chemsemse data
chem = pd.read_csv('chemsense.csv',header = 0,index_col=0)
```

```python
chem['AT'].describe()
```

```python
#Round the temperature to 0.1 degree.

chem['AT'] = round(chem['AT'],1)
```

```python
#Plot the NO2 VS temp for whole dataset
fig, ax = plt.subplots(3,1,figsize=(20,15))

for i, sensor in enumerate(chem.BAD.unique()):
    data = chem[chem['BAD']==sensor]
    data.index = pd.to_datetime(data.index)
    ax[i].scatter(data['AT'], data['NO2'])
    ax[i].set_xlabel('AT')
    ax[i].set_ylabel('no2_raw')
    ax[i].set_title(sensor+": whole dataset: NO2 VS AT")
fig.autofmt_xdate()
```

```python
#Plot the NO2 VS temp for bottom 5% value at each temprature

fig, ax = plt.subplots(3,1,figsize=(20,15))

for i, sensor in enumerate(chem.BAD.unique()):
    data = chem[chem['BAD']==sensor]
    data_bottom = data[(data['NO2'] < data.groupby('AT')['NO2'].transform(lambda x: x.quantile(0.05)))]
    ax[i].scatter(data_bottom['AT'],data_bottom['NO2'])
    ax[i].set_xlabel('AT')
    ax[i].set_ylabel('no2_raw')
    ax[i].set_title(sensor+": bottom 5% values at each temp: NO2 VS AT")
fig.autofmt_xdate()
```
